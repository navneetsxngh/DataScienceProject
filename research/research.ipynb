```python
## Initializing a Dictionary
example = {"Key1" : "Value1", "Key2" : "Value2"}

## To call any value in a dictionary
example["Key1"]
```

```python
example.key1   ## This is not supported
```

```python
## We use ConfigBox to call any value from dict
from box import ConfigBox

example = ConfigBox({"Key1": "Value1", "Key2": "Value2"})
example.Key1
```

```python
## Ensure Annotation

def get_products(x:int, y:int) -> int:
    return x * y

get_products(2, 2)
```

```python
get_products(2 , "4")
```

```python
get_products(6 , "6")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_products(x:int, y:int) -> int:
    return x * y
```

```python
get_products(3,4)
```

```python
get_products(2, "6")
```
